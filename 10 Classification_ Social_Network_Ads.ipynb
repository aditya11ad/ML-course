# Logistic Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
dataset
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

## Training the Logistic Regression model on the Training set

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

## Predicting a new result

```python
print(classifier.predict(sc.transform([[30,87000]])))
```

## Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
classifier.predict_proba(X_test)
```

### Precision:

accuracy of the positive predictions

**Precision:** 

The **precision** will be "how many are correctly classified among that class"

# Recall (or sensitivity or the true positive rate) :

ratio of positive instances that are correctly detected by the classifier

```python
recall_score(y_test, y_pred)
```

```python
# == TP / (TP + FN )
24 / (24 + 8 )
```

**Recall:**
The **recall** means "how many of this class you find over the whole number of element of this class"

24 out of 32

it detects 75% of the customers who Purchased.

***
The **f1-score** is the harmonic mean between precision & recall

The **support** is the number of occurence of the given class in your dataset (so you have 68 of class 0 and 32 of class 1, which is a really well balanced dataset

#### OR

```python
def plot_roc(y_trues, y_preds, labels, x_max=1.0):
    fig, ax = plt.subplots()
    for i, y_pred in enumerate(y_preds):
        y_true = y_trues[i]
        fpr, tpr, thresholds = roc_curve(y_true, y_pred)
        auc = roc_auc_score(y_true, y_pred)
        ax.plot(fpr, tpr, label='%s; AUC=%.3f' % (labels[i], auc), marker='o', markersize=1)

    ax.legend()
    ax.grid()
    ax.plot(np.linspace(0, 1, 20), np.linspace(0, 1, 20), linestyle='--')
    ax.set_title('ROC curve')
    ax.set_xlabel('False Positive Rate')
    ax.set_xlim([-0.01, x_max])
    _ = ax.set_ylabel('True Positive Rate')
```

```python
plot_roc(
    [y_test]*2,
    [ypred_prob,y_scores_forest],
    ['Logistic Regression','Random Forest']
)
```
